```python
from deepface import DeepFace
import cv2
import os

# --- CONFIG ---
dataset_path = r"C:\Users\Kamal Rawat\Documents\dataset\faces_data"
test_image_path = r"C:\Users\Kamal Rawat\Downloads\aila.webp"
model = "Facenet"  # You can also try "ArcFace" or "VGG-Face"

# ‚ö° STEP 1: Build or load cached embeddings
print("üîç Searching for matches...")

try:
    df = DeepFace.find(
        img_path=test_image_path,
        db_path=dataset_path,
        model_name=model,
        enforce_detection=False,
        detector_backend='opencv',
        silent=True,
        distance_metric='euclidean_l2'
    )
except Exception as e:
    print("‚ùå Error during face search:", e)
    df = None

# ‚ö° STEP 2: Display top match
if df is not None and len(df) > 0 and not df[0].empty:
    best_match = df[0].iloc[0]
    identity = best_match['identity']
    distance = best_match['distance']

    # Extract name from folder
    person_name = os.path.basename(os.path.dirname(identity))

    print(f"‚úÖ Match found: {person_name}")
    print(f"üì∏ Image Path: {identity}")
    print(f"üîπDistance: {distance:.3f}")

    # Show both images
    img_test = cv2.imread(test_image_path)
    img_match = cv2.imread(identity)

    if img_test is not None and img_match is not None:
        cv2.imshow("Input Image", cv2.resize(img_test, (400, 400)))
        cv2.imshow(f"Matched Person: {person_name}", cv2.resize(img_match, (400, 400)))
        cv2.waitKey(0)
        cv2.destroyAllWindows()
else:
    print("‚ùå No match found in dataset.")
```

```python
import os
import cv2
import numpy as np
import face_recognition

# --- CONFIG ---
dataset_path = r"C:\Users\Kamal Rawat\Documents\dataset\faces_data"

print("üì¶ Loading known faces...")

known_encodings = []
known_names = []

# Step 1: Load dataset only once
for person_name in os.listdir(dataset_path):
    person_folder = os.path.join(dataset_path, person_name)
    if not os.path.isdir(person_folder):
        continue

    for img_name in os.listdir(person_folder):
        img_path = os.path.join(person_folder, img_name)
        image = face_recognition.load_image_file(img_path)
        encodings = face_recognition.face_encodings(image)
        if len(encodings) > 0:
            known_encodings.append(encodings[0])
            known_names.append(person_name)

print(f"‚úÖ Loaded {len(known_names)} known faces")

# Step 2: Start webcam
video = cv2.VideoCapture(0)
print("üé• Starting camera... (Press 'Q' to exit)")

while True:
    ret, frame = video.read()
    if not ret:
        break

    # Resize frame for faster processing
    small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)
    rgb_small_frame = cv2.cvtColor(small_frame, cv2.COLOR_BGR2RGB)

    # Detect faces
    face_locations = face_recognition.face_locations(rgb_small_frame)
    face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # Compare to known faces
        distances = face_recognition.face_distance(known_encodings, face_encoding)
        best_match_index = np.argmin(distances)

        if distances[best_match_index] < 0.45:
            name = known_names[best_match_index]
        else:
            name = "Unknown"

        # Scale back up
        top, right, bottom, left = top * 4, right * 4, bottom * 4, left * 4

        # Draw box and name
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(frame, name, (left, top - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    cv2.imshow("‚ö° Fast Face Recognition", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
